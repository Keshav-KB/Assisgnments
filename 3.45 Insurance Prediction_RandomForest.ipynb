```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset1=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset1.columns
```

```python
independent=dataset1[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependent=dataset1[['charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators = 1000, max_depth = 50, max_features = "sqrt", bootstrap = "true",min_samples_leaf = 1)
```

```python
regressor.fit(X_train,Y_train) #Fit method does the training and generates the model once training completes
```

```python
Y_predict=regressor.predict(X_test)
#Y_predict
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(Y_test,Y_predict)
```

```python
r_score
```

```python
import pickle
filename="finalised_model_Forest.sav"
pickle.dump(regressor,open(filename,'wb'))
```
